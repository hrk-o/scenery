```python
import random
```

```python
shizuoka_imgs = pd.read_csv('./shizuoka_imgs.csv',header=None, names=['id', 'lat', 'lon'])
shizuoka_imgs
```

```python
shizuoka_duration = pd.read_csv('./shizuoka_duration.csv',header=None)
shizuoka_duration
```

```python
shizuoka_startduration = pd.read_csv('./shizuoka_startduration.csv',header=None)
shizuoka_startduration
```

```python
images = len(shizuoka_duration)
duration_matrix = [[0 for i in range(images+1)]for j in range(images+1)]
for i in range(len(shizuoka_duration)):
    for j in range(len(shizuoka_duration.iloc[i,:])):
        if i > j:
            duration_matrix[i][j] = int(shizuoka_duration.iloc[i,j])
            duration_matrix[j][i] = int(shizuoka_duration.iloc[i,j])
    for j in range(len(shizuoka_startduration.iloc[0,:])):
        duration_matrix[j][images] = int(shizuoka_startduration.iloc[0,j])
        duration_matrix[images][j] = int(shizuoka_startduration.iloc[0,j])
        
duration_matrix
pd_duration_matrix = pd.DataFrame(duration_matrix,
                  index=[i for i in range(images+1)],
                  columns=[i for i in range(images+1)])
pd_duration_matrix
```

```python
startcell = 50
images = 94
selectimages = 3 #大体何拠点くらいを選ばせたいか
#初期集団の生成
cells = [[1 if random.random() <= selectimages/images else 0 for i in range(images)] for j in range(startcell)]
```

```python
#適応度評価
fitness = []
for i in range(startcell):
    indexs = []
    score = -999
    for j in range(len(cells[i])):
        if cells[i][j] == 1:
            indexs.append(j)
    if len(indexs) == 0:
        fitness.append(score)
    else:
        #距離
        dist_score = 0
        dist = pd_duration_matrix.iloc[indexs + [images],indexs + [images]].values
        route = ortools_vrp(len(dist), dist, limit_time=1)[0]
        for r in range(len(route)):
            if r != 0:
                dist_score += dist[route[r-1]][route[r]]
        #print(route)
        if (dist_score/3600 < 2) or (dist_score/3600 > 4):#何時間以上、何時間以内か計算
            dist_score = -999
        else:
            dist_score = len(indexs)
        #AHP
        score = dist_score
        fitness.append(score)#距離+AHP結果
    print(indexs,score)
#fitness 適応度
```
