```python
import pandas as pd
import numpy as np
import cv2
from PIL import Image

# インライン表示
%matplotlib inline
```

```python
name = '255783806'
images = pd.read_csv('./image-data_colors/{}.csv'.format(name),header=None)
images
```

```python
colors = np.array([[255, 0, 0], [255, 127, 0], [255, 255, 0], [127, 255, 0], [0, 255, 0], [0, 255, 127], [0, 255, 255], [0, 127, 255], [0, 0, 255], [127, 0, 255], [255, 0, 255], [255, 0, 127]])
colors_index = [8, 7,6, 5, 4, 3, 2, 1 ,0 ,11 ,10, 9]
```

```python
filepath = './image-data/{}.jpg'.format(name)
bgr_array = cv2.imread(filepath)
```

```python
for i in range(len(bgr_array)):
    for j in range(len(bgr_array[i])):
        bgr_array[i][j] = list(colors[colors_index[images.iloc[i,j]]])
```

```python
display(Image.fromarray(bgr_array))
```
