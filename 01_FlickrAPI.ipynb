```python
# 必要なライブラリ
from flickrapi import FlickrAPI
import folium
import json
import pandas as pd
import time
import traceback
from urllib.request import urlretrieve
import os
import csv

def get_photos(url, filepath):
    urlretrieve(url, filepath)
    time.sleep(1)

# Flickr APIを利用するための情報を記載
key = "******"
secret = "******"

keyword = '静岡'

# Flickr APIを使って写真の情報を取得
# 今回は写真のタイトルに"fish"が関連しているものを５００件を上限に取得
# 位置情報は、extrasでgeoを指定
flickr = FlickrAPI(key, secret, format='parsed-json')

result  = flickr.photos.search(
    text = keyword,
    per_page = 500,
    media = 'photos',
    sort = 'relevance',
    safe_seach = 1,
    extras = 'url_q,url_o, licence, geo,date_taken',
)

photos = result['photos']['photo']

# 地図の中心を、写真に保存されている緯度経度の平均を出すため
lat_sum, lon_sum = 0,0

for photo in photos:
  lat_sum += float(photo['latitude'])
  lon_sum += float(photo['longitude'])

lat_mean = lat_sum / len(photos)
lon_mean = lon_sum / len(photos)

# 元となる地図を作成
map = folium.Map(location=[lat_mean, lon_mean], zoom_start=3)

infophotos = []

# 写真に保存されている位置にピンをたて、そのピンに写真を載せる
for photo in photos:
  if 'url_o' in list(photo) :
    filepath = './image-data/'+ photo['id'] + '.jpg'
    url_o = photo['url_o']
    get_photos(url_o, filepath)
    infophotos.append([photo['id'], float(photo['latitude']), float(photo['longitude']), photo['datetaken'], photo['title']])#20230418風景追加
  folium.Marker(
      location=[float(photo['latitude']), float(photo['longitude'])],
      popup="<p>{}<br /><img src='{}'</p>".format(photo['title'],photo['url_q'])
  ).add_to(map)

# 地図を描画
map

with open('image.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerows(infophotos)
```
