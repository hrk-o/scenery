## 取得した画像をmapに表示

```python
import pandas as pd
import folium
from PIL import Image
import csv
```

```python
images = pd.read_csv('./image.csv',header=None, names=['id', 'lat', 'lon'])
images
```

```python
# 元となる地図を作成
map = folium.Map(location=[images[images.lat != 0].mean()['lat'], images[images.lat != 0].mean()['lon']], zoom_start=8)
map
```

```python
# 写真に保存されている位置にピンをたて、そのピンに写真を載せる
for i in range(len(images)):
  if images.iloc[i,2] != 0:    
    filepath = './image-data/{}.jpg'.format(images.iloc[i,0])
    #img = Image.open(filepath)
    folium.Marker(
      location=[float(images.iloc[i,1]), float(images.iloc[i,2])],
      popup="<p>{}<br /><img src='{}' width='300' </p>".format(images.iloc[i,0],filepath)
    ).add_to(map)
```

## 取得した画像をmapに表示（静岡県のみとする=目視で確認）

```python
# 元となる地図を作成
map = folium.Map(location=[images[images.lat != 0].mean()['lat'], images[images.lat != 0].mean()['lon']], zoom_start=8)
# 写真に保存されている位置にピンをたて、そのピンに写真を載せる
photos_num = 0
shizuoka_imgs = []
for i in range(len(images)):
  if (images.iloc[i,2] != 0) and (images.iloc[i,0] != 9433346527):
    photos_num += 1
    shizuoka_imgs.append(list(images.iloc[i,:]))
    filepath = './image-data/{}.jpg'.format(images.iloc[i,0])
    #img = Image.open(filepath)
    folium.Marker(
      location=[float(images.iloc[i,1]), float(images.iloc[i,2])],
      popup="<p>{}<br /><img src='{}' width='300' </p>".format(images.iloc[i,0],filepath)
    ).add_to(map)
print(photos_num)
map
```

```python
with open('./shizuoka_imgs.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(shizuoka_imgs)
```
